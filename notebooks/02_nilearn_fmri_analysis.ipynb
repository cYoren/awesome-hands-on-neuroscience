```python
import nilearn.datasets as nld
import nilearn.plotting as nlp
from nilearn.maskers import NiftiMasker
from nilearn.image import mean_img
from nilearn.connectome import ConnectivityMeasure
import matplotlib.pyplot as plt
import numpy as np

# Load Haxby dataset (fMRI while viewing faces, houses, cats, bottles...)
haxby = nld.fetch_haxby()
print('Conditions:', set(open(haxby.session_target[0]).read().split()))

# Plot mean BOLD image
mean_bold = mean_img(haxby.func[0])
nlp.plot_epi(mean_bold, title='Mean BOLD \u2014 Haxby visual fMRI')
plt.show()
```

```python
# Plot the 20 resting-state networks from Smith 2009
smith = nld.fetch_atlas_smith_2009()
nlp.plot_prob_atlas(smith.maps, title='Smith 2009: 20 RSNs', colorbar=True)
plt.show()
```

```python
# Build a functional connectome from ADHD resting-state data
adhd = nld.fetch_adhd(n_subjects=1)
msdl = nld.fetch_atlas_msdl()
# MSDL is a probabilistic atlas (4D) \u2014 use NiftiMapsMasker, not NiftiLabelsMasker
from nilearn.maskers import NiftiMapsMasker
atlas_masker = NiftiMapsMasker(maps_img=msdl.maps, standardize='zscore_sample')
time_series = atlas_masker.fit_transform(adhd.func[0])
corr = ConnectivityMeasure(kind='correlation')
matrix = corr.fit_transform([time_series])[0]
nlp.plot_matrix(matrix, labels=msdl.labels, colorbar=True,
                vmax=0.8, vmin=-0.8, title='MSDL functional connectome (ADHD200)')
plt.show()
```
